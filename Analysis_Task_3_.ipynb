```python
# ==========================================
# Data Science Internship – Task 3
# Data Processing and Analysis
# ==========================================


# ============================================================
# 1️⃣ Employee Performance Bonus Eligibility
# ============================================================

print("----- Problem 1: Employee Performance Bonus Eligibility -----")

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest_score = max(employees.values())

# Handle ties
top_performers = [name for name, score in employees.items() if score == highest_score]

print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")

```

```python
# ============================================================
# 2️⃣ Search Query Keyword Analysis
# ============================================================

print("\n----- Problem 2: Search Query Keyword Analysis -----")

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation (if any)
import string
for punct in string.punctuation:
    query = query.replace(punct, "")

words = query.split()

# Count frequency
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Display only words searched more than once
repeated_keywords = {word: count for word, count in frequency.items() if count > 1}

print(repeated_keywords)

```

```python
# ============================================================
# 3️⃣ Sensor Data Validation
# ============================================================

print("\n----- Problem 3: Sensor Data Validation -----")

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:  # even number = valid reading
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# ============================================================
# 4️⃣ Email Domain Usage Analysis
# ============================================================

print("\n----- Problem 4: Email Domain Usage Analysis -----")

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total_emails = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")

```

```python
# ============================================================
# 5️⃣ Sales Spike Detection
# ============================================================

print("\n----- Problem 5: Sales Spike Detection -----")

sales = [1200, 1500, 900, 2200, 1400, 3000]

average_sales = sum(sales) / len(sales)

threshold = average_sales * 1.30  # 30% above average

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
# ============================================================
# 6️⃣ Duplicate User ID Detection
# ============================================================

print("\n----- Problem 6: Duplicate User ID Detection -----")

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

user_frequency = {}

for user in user_ids:
    user_frequency[user] = user_frequency.get(user, 0) + 1

for user, count in user_frequency.items():
    if count > 1:
        print(f"{user} → {count} times")


print("\n✅ All Task 3 problems executed successfully!")
```
