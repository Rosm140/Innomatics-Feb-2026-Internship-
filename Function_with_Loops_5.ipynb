```python
# ==========================================
# Data Science Internship ‚Äì Task 5
# Function-Based Problem Solving
# ==========================================


# ============================================================
# 1Ô∏è‚É£ Social Media ‚Äì Post Engagement Analyzer
# ============================================================

def analyze_engagement(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like

    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


analyze_engagement([200, 300, 150, 400])

```

```python
# ============================================================
# 2Ô∏è‚É£ Healthcare ‚Äì Medicine Stock Alert System
# ============================================================

def check_medicine_stock(stock):
    status = "Low Stock Alert" if stock < 10 else "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)


check_medicine_stock(6)
```

```python
# ============================================================
# 3Ô∏è‚É£ Agriculture ‚Äì Rainfall Adequacy Checker
# ============================================================

def rainfall_checker(rainfall_list, required_level):
    total = 0
    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)
    status = "Adequate Rainfall" if average >= required_level else "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)


rainfall_checker([60, 70, 80, 75], 70)
```

```python
# ============================================================
# 4Ô∏è‚É£ Social Media ‚Äì Duplicate Account Detection
# ============================================================

def detect_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


detect_duplicates(["user1", "user2", "user3", "user1"])
```

```python
# ============================================================
# 5Ô∏è‚É£ Healthcare ‚Äì Appointment Eligibility Checker
# ============================================================

def appointment_eligibility(age):
    status = "Eligible" if age >= 18 else "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)


appointment_eligibility(21)

```

```python
# ============================================================
# 6Ô∏è‚É£ Agriculture ‚Äì Premium Crop Price Filter
# ============================================================

def filter_premium_crops(prices):
    premium = []
    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)


filter_premium_crops([1500, 2500, 1800, 3200])
```

```python

# ============================================================
# 7Ô∏è‚É£ System Monitoring ‚Äì Application Health Checker
# ============================================================

def check_system_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", error_count)
    print("System Status:", status)


check_system_health(7)

```

```python

# ============================================================
# 8Ô∏è‚É£ Banking ‚Äì Daily Transaction Limit Checker
# ============================================================

def transaction_limit_checker(amount):
    daily_limit = 50000
    status = "Approved" if amount <= daily_limit else "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


transaction_limit_checker(60000)
```

```python
# ============================================================
# 9Ô∏è‚É£ E-Learning ‚Äì Student Attendance Eligibility System
# ============================================================

def attendance_checker(attendance_list):
    present_days = 0
    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)


attendance_checker(["P", "P", "A", "P", "P"])
```

```python
# ============================================================
# üîü Smart Electricity Bill Analyzer
# ============================================================

def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif 500 <= bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Units Consumed:", units)
    print("Total Bill:", bill)
    print("Usage Category:", usage)


electricity_bill(250)


print("\n‚úÖ All Task 5 Problems Executed Successfully!")
```
