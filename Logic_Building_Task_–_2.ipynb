```python
# ==========================================
# Data Science Internship ‚Äì Logic Task 2
# ==========================================


# 1Ô∏è‚É£ UNIQUE WORDS IN A SENTENCE
print("----- 1. Unique Words in a Sentence -----")

sentence = "python is easy and python is powerful"

words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python

# 2Ô∏è‚É£ HIGHEST SALARY FROM EMPLOYEE DATA
print("\n----- 2. Highest Salary from Employee Data -----")

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = max(employees, key=employees.get)

print(f"Highest Salary: {highest_employee} - {employees[highest_employee]}")

```

```python
# 3Ô∏è‚É£ FIND MAXIMUM AND MINIMUM VALUES
print("\n----- 3. Find Maximum and Minimum Values -----")

numbers = [45, 22, 89, 10, 66]

max_value = numbers[0]
min_value = numbers[0]

for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

```python
# 4Ô∏è‚É£ COUNT PRODUCTS ABOVE PRICE THRESHOLD
print("\n----- 4. Count Products Above Price Threshold -----")

prices = [450, 1200, 899, 1500, 300]

count_above_1000 = 0

for price in prices:
    if price > 1000:
        count_above_1000 += 1

print("Products above 1000:", count_above_1000)
```

```python
# 5Ô∏è‚É£ CALCULATE ATTENDANCE PERCENTAGE
print("\n----- 5. Calculate Attendance Percentage -----")

attendance = ["P", "P", "A", "P", "P"]

present_days = attendance.count("P")
total_days = len(attendance)

attendance_percentage = (present_days / total_days) * 100

print("Attendance Percentage:", attendance_percentage)
```

```python
# 6Ô∏è‚É£ REMOVE DUPLICATE PHONE NUMBERS
print("\n----- 6. Remove Duplicate Phone Numbers -----")

phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

```python
# 7Ô∏è‚É£ COUNT CHARACTER FREQUENCY
print("\n----- 7. Count Character Frequency -----")

text = "python"
char_frequency = {}

for char in text:
    if char in char_frequency:
        char_frequency[char] += 1
    else:
        char_frequency[char] = 1

print("Character frequency:", char_frequency)
```

```python
# 8Ô∏è‚É£ CONVERT LIST TO TUPLE
print("\n----- 8. Convert List to Tuple -----")

sample_list = [10, 20, 30]

sample_tuple = tuple(sample_list)

print("Tuple:", sample_tuple)
```

```python
# 9Ô∏è‚É£ CHECK IF A KEY EXISTS IN DICTIONARY
print("\n----- 9. Check if Key Exists in Dictionary -----")

employee_data = {"Ravi": 75000, "Anita": 68000}

key_to_check = "Ravi"

if key_to_check in employee_data:
    print("Employee exists")
else:
    print("Employee does not exist")
```
