```python
# ==========================================
# Data Science Internship – Task 6
# Function-Based Problem Solving
# ==========================================


# ============================================================
# 1️⃣ Smart Parking Lot Management System
# ============================================================

def smart_parking(capacity, vehicle_logs):
    parked = 0
    peak = 0

    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        if parked > peak:
            peak = parked

    status = "Available" if parked < capacity else "Full"

    print("Currently Parked Vehicles:", parked)
    print("Peak Usage:", peak)
    print("Parking Status:", status)


smart_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])


```

```python
# ============================================================
# 2️⃣ Online Food Delivery Time Estimator
# ============================================================

def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5  # 5 minutes per km

    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 8

    if weather == "Rainy":
        base_time += 10
    elif weather == "Stormy":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")


estimate_delivery_time(8, "High", "Rainy")
```

```python

# ============================================================
# 3️⃣ Movie Theatre Seat Occupancy Analyzer
# ============================================================

def seat_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", f"{occupancy:.0f}%")
    print("Show Status:", status)


seat_occupancy(200, [1]*150)
```

```python
# ============================================================
# 4️⃣ Cloud Server Load Classification System
# ============================================================

def server_load(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    if average < 50:
        status = "Normal"
    elif 50 <= average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", f"{average:.0f}%")
    print("Server Status:", status)


server_load([45, 60, 70, 85, 90])
```

```python

# ============================================================
# 5️⃣ Smart Classroom Resource Usage Monitor
# ============================================================

def resource_monitor(resource_usage):
    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    energy_alert = "Yes" if overused else "No"

    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", energy_alert)


resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})


```

```python
# ============================================================
# 6️⃣ Online Event Registration Capacity Controller
# ============================================================

def event_registration(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


event_registration(100, 105)


print("\n✅ All Task 6 Problems Executed Successfully!")
```
